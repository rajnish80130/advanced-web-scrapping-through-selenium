```python
with open('smartprix.html','r',encoding='utf-8') as f:
  html = f.read()
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
len(soup.find_all('div', class_='sm-product has-tag has-features has-actions'))
```

```python
containers = soup.find_all('div', class_='sm-product has-tag has-features has-actions')
```

```python
phone_name = []
phone_price = []
spec_score = []
sim = []
processor = []
memory = []
battery = []
display = []
camera = []
and_version = []

for i in containers:
  phone_name.append(i.find('h2').text)
  phone_price.append(i.find('span',class_='price').text)
  spec_score.append(i.find('b').text)
  sim.append(i.find_all('li')[0].text)
  processor.append(i.find_all('li')[1].text)
  memory.append(i.find_all('li')[2].text)
  battery.append(i.find_all('li')[3].text)
  display.append(i.find_all('li')[4].text)
  camera.append(i.find_all('li')[5].text)
  try:
    and_version.append(i.find_all('li')[7].text)
  except:
    and_version.append(np.nan)
```

```python
len(phone_price)
```

```python
len(spec_score)
```

```python
df = pd.DataFrame({
    'model':phone_name,
    'price':phone_price,
    'rating':spec_score,
    'sim' : sim,
    'processor':processor,
    'memory':memory,
    'battery':battery,
    'display':display,
    'camera':camera,
    'android_ver':and_version
})
```

```python
df.sample(10)
```

```python
file_name = 'phone_data.csv'
```

```python
df.to_csv(file_name, index=True)
```
